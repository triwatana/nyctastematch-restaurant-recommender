```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('nyctastematch-restaurant-recommender/data/restaurants_data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
# Removing unnecessary columns
columns_to_keep = [
    'name', 'price', 'rating', 'review_count', 'CUISINE DESCRIPTION',
    'categories_clean', 'neighborhood', 
    'address', 'city', 'state', 'zip_code', 'display_phone'
]
data = data[columns_to_keep]
```

```python
# Renaming columns
data.rename(columns={'CUISINE DESCRIPTION': 'cuisine'}, inplace=True)
data.rename(columns={'categories_clean': 'categories'}, inplace=True)
```

```python
data.head()
```

```python
# Converting the values in the price column to numerical values
price_mapping = {
    '$': 1,
    '$$': 2,
    '$$$': 3,
    '$$$$': 4
}
data['price'] = data['price'].map(price_mapping)
data['price'] = data['price'].astype('Int64') 
```

```python
# Using only restaurants that are in New York
data = data[(data['city'] == 'New York') & (data['state'] == 'NY')]
```

```python
print(f"Total restaurants after filtering: {len(data)}")
print(f"\nPrice distribution:")
print(data['price'].value_counts().sort_index())
```

```python
# Dropping state and city columns since all restaurants beling to the same city and state
data = data.drop(columns = ['state','city'])
```

```python
data.head()
```

```python
# Checking for null values
data.isnull().sum()
```

```python
# Dropping the only row which has no address
data = data.dropna(subset=['address'])
```

```python
data.head()
```

```python
# Filling null values of the price column based on the cuisine and the neighborhood of the restaurant

data['price'] = data.groupby('neighborhood')['price'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

data['price'] = data.groupby('cuisine')['price'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

if data['price'].isna().sum() > 0:
    price_distribution = data['price'].value_counts(normalize=True)
    data.loc[data['price'].isna(), 'price'] = np.random.choice(
        price_distribution.index, 
        size=data['price'].isna().sum(), 
        p=price_distribution.values
    )

data['price'] = data['price'].astype(int)
```
